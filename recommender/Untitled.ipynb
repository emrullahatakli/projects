### Demographic Filtering
Filmin ortalama reytinglerini puan olarak kullanabiliriz ama bunu kullanmak yeterince adil olmayacaktÄ±r Ã§Ã¼nkÃ¼ ortalama reytingi 8,9 olan ve sadece 3 oy alan bir film, ortalama reytingi 7,8 olan ancak 40 oy alan filmden daha iyi kabul edilemez. Bu yÃ¼zden, IMDB'nin ÅŸu ÅŸekilde verilen aÄŸÄ±rlÄ±klÄ± derecelendirmesini (wr) kullanacaÄŸÄ±m: 


* v, film iÃ§in oy sayÄ±sÄ±dÄ±r;
* m, tabloda listelenmek iÃ§in gereken minimum oy sayÄ±sÄ±dÄ±r;
* R, filmin ortalama derecelendirmesidir; Ve
* C, raporun tamamÄ±ndaki ortalama oydur

```python
v = df2["vote_count"]
R = df2["vote_average"]
C = df2["vote_average"].mean()
m = df2["vote_count"].quantile(0.9)
```

quantile deÄŸerini ben belirledim. Bir filmin listede yer alabilmesi iÃ§in listedeki filmlerin en az %90Ä±ndan daha fazla oy almasÄ± gerekiyor.

```python
#listeye uygun filmleri filtreleyelim.
q_movies = df2.copy().loc[v >= m]
q_movies.shape
```

```python
#ÅŸimdi weighted rating(wr) fonksiyonunu yazalÄ±m. DatamÄ±za uygulayalÄ±m.

def weighted_rating(x, m = m, C = C):
    v = x["vote_count"]
    R = x["vote_average"]
    
    return (v / (v + m) * R) + (m / (m + v) * C)
```

```python
q_movies["score"] = q_movies.apply(weighted_rating, axis = 1)
```

```python
#top 10 filmin oy Ã¶zelliklerini Ã§Ä±karalÄ±m.
q_movies = q_movies.sort_values("score", ascending = False)

cols = ["title", "vote_count", "vote_average" , "score"]

q_movies[cols].head(10)
```

demografik filtreleme Ã§ok geneldir. 

## Content Based Filtering


Bu Ã¶neri sisteminde filmin iÃ§eriÄŸi (genel bakÄ±ÅŸ, oyuncu kadrosu, ekip, anahtar kelime, slogan vb.) diÄŸer filmlerle benzerliÄŸini bulmak iÃ§in kullanÄ±lÄ±r. ArdÄ±ndan benzer olma olasÄ±lÄ±ÄŸÄ± en yÃ¼ksek olan filmler Ã¶nerilir.

Biz verisetindeki konu aÃ§Ä±klamasÄ±na gÃ¶re benzer filmleri bulmaya Ã§alÄ±ÅŸaÄŸÄ±z.

```python
df2["overview"].head()
```

Burda kelimeleri vektÃ¶rize edeceÄŸiz. Her bir overview iÃ§in TF-IDF vektÃ¶rÃ¼nÃ¼ bulacaÄŸÄ±z. 

TF: terim frekansÄ±dÄ±r. bir kelimenin bir dÃ¶kÃ¼mandaki sÄ±klÄ±ÄŸÄ±dÄ±r. (terim Ã¶rnekleri / terim toplamÄ±)

IDF: terimi iÃ§eren belgelerin sÄ±klÄ±ÄŸÄ±dÄ±r. log(belgelerin sayÄ±sÄ± / terimli belgelerin sayÄ±sÄ±)

Her kelimenin Ã¶nemi TF * IDF yapÄ±larak bulunur.

Bu iÅŸlemlerin sonucunda daha Ã¶nce olduÄŸu gibi, her sÃ¼tunun overview sÃ¶zlÃ¼ÄŸÃ¼ndeki bir kelimeyi (en az bir belgede gÃ¶rÃ¼nen tÃ¼m kelimeler) ve her satÄ±rÄ±n bir filmi temsil ettiÄŸi bir matris verecektir.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Bir TF-IDF Vectorizer Nesnesi tanÄ±mladÄ±k. 'the', 'a' gibi tÃ¼m ingilizce stopwordleri kaldÄ±rdÄ±k.
tfidf = TfidfVectorizer(stop_words='english')

#NanlarÄ± boÅŸ stringle doldurduk.
df2['overview'] = df2['overview'].fillna('')

#Verileri uydurarak ve dÃ¶nÃ¼ÅŸtÃ¼rerek gerekli TF-IDF matrisini oluÅŸturduk.
tfidf_matrix = tfidf.fit_transform(df2['overview'])

tfidf_matrix.shape
```

Veri setimizdeki 4800 filmi tanÄ±mlamak iÃ§in 20.000'den fazla farklÄ± kelimenin kullanÄ±ldÄ±ÄŸÄ±nÄ± gÃ¶rÃ¼yoruz.

Åžimdi filmler arasÄ± benzerlikleri bulmak iÃ§in kosinÃ¼s benzerliÄŸini kullanacaÄŸÄ±z. (Cosine sim)

```python
#elimizdeki matristen dolayÄ± linear_kernel daha hÄ±zlÄ± hesaplama saÄŸlayacaktÄ±r.
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
```

```python
# Film baÅŸlÄ±ÄŸÄ±nÄ± girdi olarak alcaz ve buna benzer 10 filmi listeleyeceÄŸiz.
def get_recommendations(title, cosine_sim=cosine_sim):
    # Film adÄ±na uyan indexi alacaÄŸÄ±z.
    idx = indices[title]

    # Filmimizle diÄŸer tÃ¼m filmlerin benzerlik puanÄ±nÄ± hesaplayacaÄŸÄ±z.
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Benzerlik puanlarÄ±na gÃ¶re sÄ±ralayacaÄŸÄ±z.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # En benzer 10 filmin puanlarÄ±nÄ± alÄ±yoruz. (1'den baÅŸlamamÄ±zÄ±n sebebi ilk film zaten bizim filmimiz olacaktÄ±r.)
    sim_scores = sim_scores[1:11]

    # Film indexlerini alacaÄŸÄ±z
    movie_indices = [i[0] for i in sim_scores]

    # En Ã§ok benzeyen 10 filmi sÄ±ralayacaÄŸÄ±z.
    return df2['title'].iloc[movie_indices]
```

Gizli faktÃ¶r bir kullanÄ±cÄ±nÄ±n veya bir Ã¶ÄŸenin sahip olduÄŸu bir Ã¶zelliÄŸi veya kavramÄ± tanÄ±mlayan geniÅŸ bir fikirdir. Ã–rneÄŸin, mÃ¼zik iÃ§in gizli faktÃ¶r, mÃ¼ziÄŸin ait olduÄŸu tÃ¼r olabilir.

SVD temel olarak, her bir kullanÄ±cÄ±yÄ± ve her bir Ã¶ÄŸeyi r boyutuna sahip gizli bir uzaya eÅŸler. Bu nedenle, doÄŸrudan karÅŸÄ±laÅŸtÄ±rÄ±labilir hale geldikÃ§e, kullanÄ±cÄ±lar ve Ã¶ÄŸeler arasÄ±ndaki iliÅŸkiyi daha iyi anlamamÄ±za yardÄ±mcÄ± olur.

```python
#SVDyi uygulamak iÃ§in Surprise kÃ¼tÃ¼phanesini import edelim.

from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
reader = Reader()
ratings = pd.read_csv('ratings_small.csv')
ratings.head()
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

svd = SVD()

#RMSE bizim iÃ§in faydalÄ± olacaktÄ±r
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

yaklaÅŸÄ±k 0.90 rmse elde ediyoruz bu bizim iÃ§in yeterli olacaktÄ±r.

```python
trainset = data.build_full_trainset()
svd.fit(trainset)
```
