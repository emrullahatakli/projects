### Demographic Filtering
Filmin ortalama reytinglerini puan olarak kullanabiliriz ama bunu kullanmak yeterince adil olmayacaktır çünkü ortalama reytingi 8,9 olan ve sadece 3 oy alan bir film, ortalama reytingi 7,8 olan ancak 40 oy alan filmden daha iyi kabul edilemez. Bu yüzden, IMDB'nin şu şekilde verilen ağırlıklı derecelendirmesini (wr) kullanacağım: 


* v, film için oy sayısıdır;
* m, tabloda listelenmek için gereken minimum oy sayısıdır;
* R, filmin ortalama derecelendirmesidir; Ve
* C, raporun tamamındaki ortalama oydur

```python
v = df2["vote_count"]
R = df2["vote_average"]
C = df2["vote_average"].mean()
m = df2["vote_count"].quantile(0.9)
```

quantile değerini ben belirledim. Bir filmin listede yer alabilmesi için listedeki filmlerin en az %90ından daha fazla oy alması gerekiyor.

```python
#listeye uygun filmleri filtreleyelim.
q_movies = df2.copy().loc[v >= m]
q_movies.shape
```

```python
#şimdi weighted rating(wr) fonksiyonunu yazalım. Datamıza uygulayalım.

def weighted_rating(x, m = m, C = C):
    v = x["vote_count"]
    R = x["vote_average"]
    
    return (v / (v + m) * R) + (m / (m + v) * C)
```

```python
q_movies["score"] = q_movies.apply(weighted_rating, axis = 1)
```

```python
#top 10 filmin oy özelliklerini çıkaralım.
q_movies = q_movies.sort_values("score", ascending = False)

cols = ["title", "vote_count", "vote_average" , "score"]

q_movies[cols].head(10)
```

demografik filtreleme çok geneldir. 

## Content Based Filtering


Bu öneri sisteminde filmin içeriği (genel bakış, oyuncu kadrosu, ekip, anahtar kelime, slogan vb.) diğer filmlerle benzerliğini bulmak için kullanılır. Ardından benzer olma olasılığı en yüksek olan filmler önerilir.

Biz verisetindeki konu açıklamasına göre benzer filmleri bulmaya çalışağız.

```python
df2["overview"].head()
```

Burda kelimeleri vektörize edeceğiz. Her bir overview için TF-IDF vektörünü bulacağız. 

TF: terim frekansıdır. bir kelimenin bir dökümandaki sıklığıdır. (terim örnekleri / terim toplamı)

IDF: terimi içeren belgelerin sıklığıdır. log(belgelerin sayısı / terimli belgelerin sayısı)

Her kelimenin önemi TF * IDF yapılarak bulunur.

Bu işlemlerin sonucunda daha önce olduğu gibi, her sütunun overview sözlüğündeki bir kelimeyi (en az bir belgede görünen tüm kelimeler) ve her satırın bir filmi temsil ettiği bir matris verecektir.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Bir TF-IDF Vectorizer Nesnesi tanımladık. 'the', 'a' gibi tüm ingilizce stopwordleri kaldırdık.
tfidf = TfidfVectorizer(stop_words='english')

#Nanları boş stringle doldurduk.
df2['overview'] = df2['overview'].fillna('')

#Verileri uydurarak ve dönüştürerek gerekli TF-IDF matrisini oluşturduk.
tfidf_matrix = tfidf.fit_transform(df2['overview'])

tfidf_matrix.shape
```

Veri setimizdeki 4800 filmi tanımlamak için 20.000'den fazla farklı kelimenin kullanıldığını görüyoruz.

Şimdi filmler arası benzerlikleri bulmak için kosinüs benzerliğini kullanacağız. (Cosine sim)

```python
#elimizdeki matristen dolayı linear_kernel daha hızlı hesaplama sağlayacaktır.
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
```

```python
# Film başlığını girdi olarak alcaz ve buna benzer 10 filmi listeleyeceğiz.
def get_recommendations(title, cosine_sim=cosine_sim):
    # Film adına uyan indexi alacağız.
    idx = indices[title]

    # Filmimizle diğer tüm filmlerin benzerlik puanını hesaplayacağız.
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Benzerlik puanlarına göre sıralayacağız.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # En benzer 10 filmin puanlarını alıyoruz. (1'den başlamamızın sebebi ilk film zaten bizim filmimiz olacaktır.)
    sim_scores = sim_scores[1:11]

    # Film indexlerini alacağız
    movie_indices = [i[0] for i in sim_scores]

    # En çok benzeyen 10 filmi sıralayacağız.
    return df2['title'].iloc[movie_indices]
```

```python
get_recommendations('Avatar')
```

```python
get_recommendations('The Karate Kid')
```

## Collaborative Filtering

Yaptığımız model sadece konuları benzer filmleri yakalayabilir. Zevkleri yakalama konusunda yetisi yoktur.

Bu modeli SVD(singular value decomposition) kullanarak yapacağız. 

Kullanıcılar ve öğeler arasındaki benzerliği yakalamak için bir gizli faktör modelinden yararlanacağız. Gizli faktör bir kullanıcının veya bir öğenin sahip olduğu bir özelliği veya kavramı tanımlayan geniş bir fikirdir. Örneğin, müzik için gizli faktör, müziğin ait olduğu tür olabilir.

SVD temel olarak, her bir kullanıcıyı ve her bir öğeyi r boyutuna sahip gizli bir uzaya eşler. Bu nedenle, doğrudan karşılaştırılabilir hale geldikçe, kullanıcılar ve öğeler arasındaki ilişkiyi daha iyi anlamamıza yardımcı olur.

```python
#SVDyi uygulamak için Surprise kütüphanesini import edelim.

from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
reader = Reader()
ratings = pd.read_csv('ratings_small.csv')
ratings.head()
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

svd = SVD()

#RMSE bizim için faydalı olacaktır
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

yaklaşık 0.90 rmse elde ediyoruz bu bizim için yeterli olacaktır.

```python
trainset = data.build_full_trainset()
svd.fit(trainset)
```
